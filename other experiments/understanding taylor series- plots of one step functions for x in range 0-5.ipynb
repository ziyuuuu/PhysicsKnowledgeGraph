```python
import networkx as nx
import sys
import cs6216_taylor_series_identification as TSI
import cs6216_knowledge_graph_creation_and_expansion_experiments as KGCE
import math
import numpy as np
import pandas as pd  
from matplotlib import cm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("experimental_results_one_step.csv")
print(df[df["ans"]=="quad"])
print(df)
```

```python
#sine 

X = np.linspace(100,1000,num=10)

Y = df[df["ans"]=="sine"]["sine(X)"].values.reshape(10,10)

colors = [cm.rainbow(i) for i in np.linspace(0, 1, len(X))]

plt.figure(figsize = (10,10))
# Plot the lines using a for loop
for i in range(10):
    plt.plot(X, Y[i], label=X[i]/10, linewidth=2., color=colors[i])

plt.legend(bbox_to_anchor=(0.7, .2), loc=2, borderaxespad=0., ncol=2)
plt.grid(True, which='both')
plt.axis([0, 1000, 0, 1])


plt.show()
```
