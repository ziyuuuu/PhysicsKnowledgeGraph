```python
import networkx as nx
import sys
import cs6216_taylor_series_identification as TSI
import cs6216_knowledge_graph_creation_and_expansion_experiments as KGCE
import math
import numpy as np
import pandas as pd  
from matplotlib import cm
import matplotlib.pyplot as plt
```

```python
#sine 

X = np.linspace(-15,15,num=100)

def ts_sine(x, series_length):
    sine_taylor_series = [-1/math.factorial(i)*(i%2 != 0)*((i+1)%4==0)+1/math.factorial(i)*((i+3)%4 == 0) for i in range(series_length)]
    power_vector = TSI.create_power_matrix(x, series_length)
    return np.sum(np.multiply(power_vector, sine_taylor_series))

series_length_array = range(100,0, -10) 
colors = [cm.rainbow(i) for i in np.linspace(0, 1, len(series_length_array))]

plt.figure(figsize = (10,10))
# Plot the lines using a for loop
for i in range(len(series_length_array)):
    plt.plot(X, list(map(ts_sine,X, [series_length_array[i]]*100)), label=series_length_array[i], linewidth=2., color=colors[i])


plt.plot(X, np.sin(X), linewidth=3., color="black", label = "sine")

plt.legend(bbox_to_anchor=(0.7, .2), loc=2, borderaxespad=0., ncol=2)
plt.grid(True, which='both')
plt.axis([-15, 15, -3, 3])


plt.show()
```

```python
#cosine 

X = np.linspace(-15,15,num=100)

def ts_cosine(x, series_length):
    cosine_taylor_series = [1/math.factorial(i)*(i%4 == 0) + -1/math.factorial(i)*((i+2)%4 == 0) for i in range(series_length)]
    power_vector = TSI.create_power_matrix(x, series_length)
    return np.sum(np.multiply(power_vector, cosine_taylor_series))

series_length_array = range(100,0, -10) 
colors = [cm.rainbow(i) for i in np.linspace(0, 1, len(series_length_array))]

plt.figure(figsize = (10,10))
# Plot the lines using a for loop
for i in range(len(series_length_array)):
    plt.plot(X, list(map(ts_cosine,X, [series_length_array[i]]*100)), label=series_length_array[i], linewidth=2., color=colors[i])


plt.plot(X, np.cos(X), linewidth=3., color="black", label = "cosine")

plt.legend(bbox_to_anchor=(0.7, .2), loc=2, borderaxespad=0., ncol=2)
plt.grid(True, which='both')
plt.axis([-15, 15, -3, 3])


plt.show()
```

```python
#quad 

X = np.linspace(-15,15,num=100)

def ts_quad(x, series_length):
    quad_taylor_series = [0+1*(i==2) for i in range(series_length)]
    power_vector = TSI.create_power_matrix(x, series_length)
    return np.sum(np.multiply(power_vector, quad_taylor_series))

series_length_array = range(100,0, -10) 
colors = [cm.rainbow(i) for i in np.linspace(0, 1, len(series_length_array))]

plt.figure(figsize = (10,10))
# Plot the lines using a for loop
for i in range(len(series_length_array)):
    plt.plot(X, list(map(ts_quad,X, [series_length_array[i]]*100)), label=series_length_array[i], linewidth=2., color=colors[i])


plt.plot(X, X**2, linewidth=3., color="black", label = "quadratic")

plt.legend(bbox_to_anchor=(0.7, .2), loc=2, borderaxespad=0., ncol=2)
plt.grid(True, which='both')
plt.axis([-15, 15, -1, 10])


plt.show()
```

```python
#exponential 

X = np.linspace(-15,15,num=100)

def ts_expon(x, series_length):
    expon_taylor_series = [1/math.factorial(i)*(i!=0) for i in range(series_length)]
    power_vector = TSI.create_power_matrix(x, series_length)
    return np.sum(np.multiply(power_vector, expon_taylor_series))

series_length_array = range(10,110, 10) 
colors = [cm.rainbow(i) for i in np.linspace(0, 1, len(series_length_array))]

plt.figure(figsize = (10,10))
# Plot the lines using a for loop
for i in range(len(series_length_array)):
    plt.plot(X, list(map(ts_expon,X, [series_length_array[i]]*100)), label=series_length_array[i], linewidth=2., color=colors[i])


plt.plot(X, np.exp(X), linewidth=3., color="black", label = "exponential")

plt.legend(bbox_to_anchor=(0.7, .2), loc=2, borderaxespad=0., ncol=2)
plt.grid(True, which='both')
plt.axis([-15, 15, -10, 10])


plt.show()
```
